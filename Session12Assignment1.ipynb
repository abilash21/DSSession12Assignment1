```python

%matplotlib inline
import pandas as pd
import numpy as np

url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)


```

```python
#1. Create a pie chart presenting the male/female proportion

tt = titanic.groupby('sex')['sex'].count().reset_index(name='total_count')
print(tt)
pp=titanic.sex.value_counts(normalize=True).reset_index(name ='total_count')
print(pp)

tt.plot.pie(y='total_count' ,autopct='%1.1f%%',startangle=90, shadow=False, labels=tt['sex'], legend = 'Proportion', fontsize=14)
pp.plot.pie(y='total_count' ,autopct='%1.2f%%',startangle=90, shadow=False, labels=pp[pp.columns[0]], legend = 'Proportion', fontsize=14)
```

```python
#2.  Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
col = np.where(titanic['sex']=='male','k','r')
titanic.plot.scatter(y='age',x='fare',marker ='^',color=col)
```
